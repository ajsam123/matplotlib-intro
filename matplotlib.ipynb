# Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic graph

```python
x = [0,1,2,3,4]
y= [0,2,4,6,8]

# Resize your graph(dpi specifies pixels per inch)
plt.figure(figsize=(3,3), dpi=200)
# plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', markersize='12', markeredgecolor='#000')

#use shorthand notation
#fmt = '[color][marker][line]'
plt.plot(x,y, 'r^--', label = '2x')

## Line number two
x2 = np.arange(0,4.5,0.5)
plt.plot(x2, x2**2, label='x^2')

# Add a title(specified font parameters with font dict)
plt.title('Our first graph', fontdict={'fontname' : 'Comic Sans MS', 'fontsize': 20})

# x and y labels
plt.xlabel('X axis')
plt.ylabel('Y axis')

# X, Y axus Tickmarks (scale of the graph)
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])

# Add a legend
plt.legend()

# Save figurehead (dpi 200 is good when saving so graph has high resolution)
plt.savefig('mygraph.png', dpi=200)

# Show plot
plt.show()
```

# Bar chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels,values)

# bars[0].set_hatch('/')
# bars[1].set_hatch('o')
# bars[2].set_hatch('*')

patterns = ['/', 'o', '*']
for idx, x in enumerate(patterns):
    bars[idx].set_hatch(x)
    
    

plt.figure(figsize=(6,4))

plt.show()
```

# Real world examples

### Line graph

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(7,5))

# plt.title('Gas Prices over time(In USD)')
# plt.plot(gas.Year, gas.USA,'b.-', label='United States')
# plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada')
# plt.plot(gas.Year, gas.Australia, 'y.-', label='Canada')
# plt.plot(gas.Year, gas['South Korea'], 'g.-', label='South Korea')

# Another way to plot many values
for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker='.', label=country)
    
plt.xticks(gas.Year[::3])

plt.xlabel('Year')
plt.ylabel('US Dollars')
plt.legend()

plt.savefig('Gas_price_figure')


plt.show()
```

# Load Fifa Data

```python
fifa = pd.read_csv('fifa_data.csv')
```

```python
fifa.head()
```

# Histogram

```python
bins = [40,50,60,70,80,80,90,100]

plt.hist(fifa.Overall, bins = bins, color='red')

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill level')
plt.title('Distribution of player skills in FIFA 2019')

plt.show()
```

# Pie Chart

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count().iloc[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count().iloc[0]

labels = ['Left', 'Right']
colors = ['#abcdef', '#00e']
plt.pie([left, right], labels = labels, colors = colors, autopct='%.2f %%')

plt.title('Foot preference of fifa players')

plt.show()
```
